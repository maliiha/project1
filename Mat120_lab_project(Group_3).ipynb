### According to Biot-Savart Law, the magnetic intensity at any point due to a steady current in an infinitely long straight wire is directly proportional to the current and inversely proportional to the distance from point to wire.

## Explanation for 1.
\begin{align}
\left| dB\right|=(\frac{\mu _0}{4\pi })(\frac{Idl\,sin\theta }{r^2})\\[1em]
\Rightarrow B = \int\mathrm{d}B = \int_{-∞}^\infty \frac{\mu_0}{4\pi} \frac{I\mathrm{d}l\sin{\theta}}{r^2} \\\\[1em]
\Rightarrow B = \frac{\mu_0I}{4\pi} \int_{-∞}^\infty \frac{\mathrm{d}l\sin{\theta}}{r^2} \\[1em]
\text{Here, }\cot{\theta} = \frac{-l}{a} \\[1em]
⇒ -l = a\cot{θ} \\[1em]
\text{After differentiating, } \mathrm{d}l = a\csc^2{\theta}\,\mathrm{d}θ \\[1em]
\text{Again, }\csc{\theta} = \frac{r}{a} \\[1em]
⇒ r = a\csc{θ} \\[1em]
\text{From the figure, we can determine that} \\[1em]
\theta = 0 \text{ when } l = -∞  \\[1em]
\theta = \pi \text{ when } l = ∞ \\[1em]
\text{After substituting values in the equation, we get,} \\[1em]
\Rightarrow B = \frac{\mu_0I}{4\pi} \int_{0}^\pi \frac{(a\csc^2{\theta})\mathrm{d}\theta\,\sin{\theta}}{a^2\csc^2{\theta}} \\[1em]
\Rightarrow B = \frac{\mu_0I}{4\pi a} \int_{0}^\pi \sin{\theta}\,\mathrm{d}\theta \\[1em]
\end{align}

```python
#solution 1
#importing library function
import sympy as smp
#declaring variable
x = smp.symbols("x")
#initiating constants
m = 4*smp.pi*10**(-7)
a = 1.5
i = 80
#declaring function
f = smp.sin(x)
#integrating function
intf = smp.integrate(f, (x, 0, smp.pi))
#using formula
b = (m*i)/(4*smp.pi*a)*intf
#printing result
print("Magnetic field = ", b, "T", sep = "")
```

## Explanation for 2.

\begin{align}
\text{As the coil is circular, the length of the conductor is equal to the circumference of the coil.} \\[1em]
\left| dB\right|=(\frac{\mu _0}{4\pi })(\frac{Idl\,sin\theta }{r^2})\\[1em]
\Rightarrow B = \int\mathrm{d}B = \int_{0}^{2\pi r} \frac{\mu_0}{4\pi} \frac{I\mathrm{d}l\sin{\theta}}{r^2} \\\\[1em]
\text{As $r$ is always perpendicular to $\textrm{d}l$, } \\[1em]
\Rightarrow B = \int_{0}^{2\pi r} \frac{\mu_0}{4\pi} \frac{I\mathrm{d}l\sin{90°}}{r^2} \\\\[1em]
\Rightarrow B = \frac{\mu_0I}{4\pi r^2} \int_{0}^{2\pi r} \mathrm{d}l \\[1em]
\text{For $N$ number of turns, } \\[1em]
\Rightarrow B = \frac{\mu_0NI}{4\pi r^2} \int_{0}^{2\pi r} \mathrm{d}l \\[1em]
\end{align}

```python
#solution 2
#importing library function
import sympy as smp
import math
#declaring variable
x = smp.symbols("x")
#initiating constants
m = 4*math.pi*10**(-7)
n = 240
r = 6.28*10**(-4)
i = 5
#declaring function
f = 1
#integrating function
intf = smp.integrate(f, (x, 0, 2*math.pi*r))
#using formula
b = (m*n*i)/(4*math.pi*r**2)*intf
#printing result
print("Magnetic field = ", b, "T", sep = "")
```

## Explanation for 3.
The Lorentz Force is the force on a charged particle due to electric and magnetic fields. A charged particle in an electric field will always feel a force due to this field, of magnitude $F=qE$. A charged particle in a magnetic field will only feel a force due to the magnetic field if it is moving with a component of its velocity perpendicular to the field. If we denote Lorentz Force, charge of the particle, velocity of the particle and the magnetic field as $\overrightarrow{F_m}, q, \overrightarrow{v}$ and $\overrightarrow{B}$, we get,
\begin{align}
  \overrightarrow{F_m} = q(\overrightarrow{v}×\overrightarrow{B}) \\[1em]
  \textsf{The value of this force is, } F_m = qvB\sin{\theta}
\end{align}

Screenshot 2023-04-18 172426.png

To calculate the velocity from a position function, we can simply differentiate the function with respect to time.
\begin{align}
  v(t) = \dfrac{dx}{dt}
\end{align}

```python
#solution 3
#importing library function
import sympy as smp
#declaring variable
t = smp.symbols("t")
#declaring function
x = t**3 - 2*t**2 + t
#initiating constants
q = 1.6*10**(-19)  #charge of an electron
b = 1.06666666666667*10**(-5)  #magnetic field acquired from 1
#calculating velocity via differentiation
v = smp.diff(x, t).subs(t, 900)  #velocity of the electron
#calculation
fm = q*v*b*smp.sin(smp.pi/2)
#printing result
print("Lorentz Force = ", fm, "N", sep = "")
```

## Explanation for 4.
If potential energy, kinetic energy, amplitude, displacement from equilibrium and positive constant for a simple harmonic motion are denoted as $E_p,\,E_k,\,A,\,x$ and $k$,
\begin{align}
  E_p &= \frac{1}{2}kx^2 \\[1em]
  E_k &= \frac{1}{2}k(A^2-x^2) \\[1em]
  \textsf{And the total mechanical energy, }E &= E_p + E_k \\[1em]
  ⇒ E &= \frac{1}{2}kA^2 \\[1em]
\end{align}
According to the graph, the total mechanical energy is always the same. The potential and kinetic energy equalize themselves to keep the mechanical energy constant.

```python
#solution 4
#importing library function
import matplotlib.pyplot as plt
import sympy as smp
import numpy as np
#declaring variable
t = smp.symbols("t")
#initiating constants
k = 10  #spring constant
#declaring function
s = t**3 - 2*t**2 + t
#substituting the value of time to calculate amplitude
a = s.subs(t, 3)
#displacement from equilibrium
x = np.arange(-a, a, 0.1)
#calculating energy
ep = 0.5*k*x**2
ek = 0.5*k*(a**2-x**2)
#plotting the graph
plt.plot(x, ep, label = "Potential energy")
plt.plot(x, ek, label = "Kinetic energy")
plt.xlabel("Displacement")
plt.ylabel("Energy")
plt.legend(loc = "upper right")
plt.show()
```
